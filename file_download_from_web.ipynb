```python
#import requests
#from bs4 import BeautifulSoup
from selenium import webdriver
#from webdriver_manager.chrome import ChromeDriverManager
#from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.jpmorganchase.com/ir/quarterly-earnings")
```

```python
print(driver.title)
```

```python
element = driver.find_element(By.PARTIAL_LINK_TEXT,"1Q25 Earnings Press Release")
```

```python
print(element.get_attribute("href"))
```

```python
elements = driver.find_elements(By.PARTIAL_LINK_TEXT,"2025")
```

```python
for element in elements:
    print(element.text)
```

```python
elements = driver.find_elements(By.TAG_NAME,"button")
for element in elements:
    print(element.text)
```

```python
elements = driver.find_elements(By.TAG_NAME,"a")
for element in elements:
    print(element.text, element.get_attribute("href"))
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//a[contains(@href, '2025')]"))
)

print(element.text)

#elements  = driver.find_elements(By.XPATH, "//div[@class='document-list-item']//child::*")
#for e in elements:
#    print(e.text)
```

```python
input_y = "2024"
input_q = "1Q"
elements  = driver.find_elements(By.XPATH, "//a[contains(@href, " + input_y + ")]")
for e in elements:
    if len(e.get_attribute("text").strip('\r\n\t ')) > 1 and input_q in e.get_attribute("text"):
        print(e.get_attribute("text").strip('\r\n\t '), e.get_attribute("href"))
```
